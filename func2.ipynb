```python
import random
from matplotlib import pyplot as plt
```

```python
def biasedRandom():
    return random.random() > 0.7
```

```python
x = list(range(2))
y = [0 for i in range(2)]

for i in range(1000):
    if (biasedRandom()):
        y [0] +=1
    else:
        y [1] +=1


plt.plot(x, y)
plt.show()
```

```python
def uniSelect():
    while True:
        l = [biasedRandom() for i in range(2)]
        if l[0] and not(l[1]):
            return True
        if l[1] and not(l[0]):
            return False
        
```

```python
x = list(range(2))
y = [0 for i in range(2)]

for i in range(10000):
    if (uniSelect()):
        y [0] +=1
    else:
        y [1] +=1

fig, ax = plt.subplots(figsize = (10, 10))
plt.ylim([0, 10000])
plt.plot(x, y)
plt.show()
```
