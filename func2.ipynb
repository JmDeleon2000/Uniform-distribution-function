```python
from random import uniform
from matplotlib import pyplot as plt
```

```python
def biasedRandom():
    return uniform(-1000, 1000) > 500
```

```python
x = list(range(2))
y = [0 for i in range(2)]

for i in range(1000):
    if (biasedRandom()):
        y [0] +=1
    else:
        y [1] +=1


plt.plot(x, y)
plt.show()
```

```python
def uniSelect():
    Y = 0
    N = 0
    for i in range(1000):
        if biasedRandom():
            Y +=1
        if biasedRandom():
            N +=1
    return Y >= N
```

```python
x = list(range(2))
y = [0 for i in range(2)]

for i in range(10000):
    if (uniSelect()):
        y [0] +=1
    else:
        y [1] +=1

fig, ax = plt.subplots(figsize = (10, 10))
plt.ylim([0, 10000])
plt.plot(x, y)
plt.show()
```
