```python
from random import uniform
from matplotlib import pyplot as plt
```

```python
def myRandom():
    return uniform(-1000, 1000) > 0
```

```python
x = list(range(2))
y = [0 for i in range(2)]

for i in range(1000):
    if (myRandom()):
        y [0] +=1
    else:
        y [1] +=1


plt.plot(x, y)
plt.show()
```

```python
def uniformRange(a, b):
    pool = list(range(a, b+1))
    while (len(pool) > 1):
        newpool = []
        while (len(newpool) < 1):
            newpool = [i for i in pool if myRandom()]
        pool = newpool
    return pool[0]
```

```python
for i in range(10):
    print(uniformRange(0, 10))
```

```python
x = list(range(10))
y = [0 for i in range(10)]

for i in range(1000000):
    y[uniformRange(0, 9)] +=1

fig, ax = plt.subplots(figsize = (10, 10))
plt.plot(x, y)

plt.ylim([0, 200000])

plt.show()
```
